```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
x1 = np.array([1,2,3,4,5,6])
x2 = np.array([7.5, 11, 16, 18, 20, 26])
```

```python
plt.scatter(x1, x2)
plt.show()
```

```python
x = np.c_[x1, x2]
x
```

```python
pca = PCA()
X_transformed = pca.fit_transform(x)
X_transformed
```
