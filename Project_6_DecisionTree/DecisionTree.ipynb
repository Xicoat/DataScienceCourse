```python
import numpy as np
import pandas as pd

def entropy(l):
    n = len(l)
    total = sum(l)
    h = 0
    for i in l:
        if i == 0:
            h -= 0
        else:
            h -= (i/total)*np.log2(i/total)
    return (h)

def expectedInf(l):
    total = 0
    for i in l:
        total += sum(i)
    inf = 0
    for i in l:
        h = entropy(i)
        inf += (sum(i)/total)*h
    return (inf)


def f_y(y):
    y = pd.Series([1 if i == 'Yes' else 0 for i in y])
    n = len(y)
    a = sum(y)
    b = n-a
    if (b != 0):
        print(f"Count of 0 = {b}")
    if (a !=  0):
        print(F"Count of 1 = {a}")
    return ([a,b])

# No cambiar
def f_X(X, y, param = None):
    _y = pd.Series([True if i == 'Yes' else False for i in y])
    if (param != None):
        mask = (X == param)
        X = X[mask]
        _y = _y[mask]
    l = []
    a = X.value_counts()
    b = X[_y].value_counts()
    for i in a.index:
        if i in b:
            l.append([b[i], a[i]-b[i]])
    return(l)

def getGain(X, y):
    initial_entropy = entropy(f_y(y))
    maxGain = float('-inf')
    param = ''
    for i in X:
        ex_Inf = expectedInf(f_X(X[i], y))
        gain = initial_entropy - ex_Inf
        if (gain > maxGain):
            maxGain = gain
            param = i
    return param, maxGain, initial_entropy

def split(X, y, feature = None, category = None):
    if feature == None:
        return 0
    mask = X[feature] == category
    t_y = y[mask].reset_index(drop=True)
    t_X = X[mask].reset_index(drop=True)
    t_X = t_X.drop(feature, axis = 1)
    return t_X, t_y

def decision(X, y, lv = 0):
    print(f"\nLevel {lv}")
    feat, gain, entropy = getGain(X, y)
    print(f"Current Entropy at is = {entropy}")
    if (entropy <= 0):
        print("Reached leaf Node")
        return
    print(f"Splitting on feature = {feat} with gain ratio {gain}")
    sub_cat = X[feat].unique()
    for i in sub_cat:
        tiny_X, tiny_y = split(X, y, feature = feat, category = i)
        decision(tiny_X, tiny_y, lv = lv+1)
```
