#### Average Funding (Without Replacement)

Problem Statement :
We want to find out the average amount of funding given to startups which are in either Bangalore or in New Delhi.
For this - rather than considering all the startups, take a sample of size 50 (without replacement). Then find the average amount of funding from this sample and calculate the Sampling Error.
Sampling Error = Average(Population) - Average(Sample)

```python
import numpy as np
import pandas as pd

data = pd.read_csv("DataSets/StartupFunding.csv", header = 0)
data.head()
df = data

# AmountInUSD to numeric
df.dropna(subset = ['AmountInUSD'], inplace = True)
df['AmountInUSD'] = df['AmountInUSD'].str.replace(',', '')
df["AmountInUSD"] = pd.to_numeric(df["AmountInUSD"], errors="coerce")

# Delhi to New Delhi
df.dropna(subset = ['CityLocation'], inplace = True)
df.loc[:,'CityLocation'] = df['CityLocation'].str.replace(r'\bDelhi\b', 'New Delhi')
df.loc[:,'CityLocation'] = df['CityLocation'].str.title()

# Small dataset
df_cities = df[['CityLocation', 'AmountInUSD']]
df_cities.head()
```

```python
n = 50
sample_50 = df_cities.sample(n)
avgSample = sample_50['AmountInUSD'].mean()
avgPopulation = df_cities['AmountInUSD'].mean()
samplingError = avgPopulation - avgSample
print('Sample mean\t\t', avgSample)
print('Population mean\t', avgPopulation)
print('Sampling Error\t', samplingError)
```

### Average Funding (With Replacement)
Problem Statement :
We want to find out the average amount of funding given to startups which are in either Bangalore or in New Delhi.
For this - rather than considering all the startups, take a sample of size 75 (with replacement). Then find the average amount of funding from this sample and calculate the Sampling Error.

```python
import numpy as np
import pandas as pd

data = pd.read_csv("DataSets/StartupFunding.csv", header = 0)
data.head()
df = data

# AmountInUSD to numeric
df.dropna(subset = ['AmountInUSD'], inplace = True)
df['AmountInUSD'] = df['AmountInUSD'].str.replace(',', '')
df["AmountInUSD"] = pd.to_numeric(df["AmountInUSD"], errors="coerce")

# Delhi to New Delhi
df.dropna(subset = ['CityLocation'], inplace = True)
df.loc[:,'CityLocation'] = df['CityLocation'].str.replace(r'\bDelhi\b', 'New Delhi')
df.loc[:,'CityLocation'] = df['CityLocation'].str.title()

# Small dataset
df_cities = df[['CityLocation', 'AmountInUSD']]
df_cities = df_cities[(df_cities['CityLocation'] == 'New Delhi') | (df['CityLocation'] == 'Bangalore')]
df_cities.head()
```

```python
n = 75
sample = df_cities.sample(n, replace=True)
avgSample = sample['AmountInUSD'].mean()
avgPopulation = df_cities['AmountInUSD'].mean()
samplingError = avgPopulation - avgSample
print('Sample mean ', avgSample)
print('Population mean ', avgPopulation)
print('Sampling Error ', samplingError)
```

#### Stratified Sampling
Problem Statement :
We want to find out the maximum amount of funding given to any startup in following regions -
Bangalore 
NCR (which includes New Delhi, Gurgaon and Noida)
Mumbai 
Pune 
Hydreabad 
So create your population data first by extracting those startups which are in given cities.
Then rather than using SRS for extracting sample, create 5 different stratas based on 5 regions and then apply stratified sampling on these 5 stratas to extract a sample of size 100 startups (taking 20 values from each strata)
Find and print Sampling error.

```python
import numpy as np
import pandas as pd

data = pd.read_csv("DataSets/StartupFunding.csv", header = 0)
data.head()
df = data

# AmountInUSD to numeric
df.dropna(subset = ['AmountInUSD'], inplace = True)
df['AmountInUSD'] = df['AmountInUSD'].str.replace(',', '')
df["AmountInUSD"] = pd.to_numeric(df["AmountInUSD"], errors="coerce")

# Delhi to New Delhi
df.dropna(subset = ['CityLocation'], inplace = True)
df.loc[:,'CityLocation'] = df['CityLocation'].str.replace(r'\bDelhi\b', 'New Delhi')
df.loc[:,'CityLocation'] = df['CityLocation'].str.title()

# Small dataset
df_cities = df[['CityLocation', 'AmountInUSD']]

# group NCR cities
replace_dict = {
    'New Delhi':'NCR',
    'Gurgaon':'NCR',
    'Noida':'NCR'
}
df_cities.loc[:,'CityLocation'] = df_cities['CityLocation'].replace(replace_dict)

cities = ['Bangalore','NCR','Mumbai','Pune','Hyderabad']
df_cities = df_cities[df_cities['CityLocation'].isin(cities)]

df_cities.head()
```

```python
cities = ['Bangalore','NCR','Mumbai','Pune','Hyderabad']

df_strat = pd.DataFrame()
for i in cities:
    df_temp = df_cities[df_cities['CityLocation'] == i]
    print(df_temp.size)
    df_temp = df_temp.sample(20)
    df_strat = pd.concat([df_strat, df_temp], ignore_index=True)
avgSample = df_strat['AmountInUSD'].mean()
avgPopulation = df_cities['AmountInUSD'].mean()
samplingError = avgPopulation - avgSample
print('Sample mean ', avgSample)
print('Population mean ', avgPopulation)
print('Sampling Error ', samplingError)
```

### Proportional Stratified Sampling
Problem Statement :
We want to find out the maximum amount of funding given to any startup in following regions -
Bangalore 
NCR (which includes New Delhi, Gurgaon and Noida)
Mumbai 
Pune 
Hydreabad 
So create your population data first by extracting those startups which are in given cities.

```python
import numpy as np
import pandas as pd

data = pd.read_csv("DataSets/StartupFunding.csv", header = 0)
data.head()
df = data

# AmountInUSD to numeric
df.dropna(subset = ['AmountInUSD'], inplace = True)
df['AmountInUSD'] = df['AmountInUSD'].str.replace(',', '')
df["AmountInUSD"] = pd.to_numeric(df["AmountInUSD"], errors="coerce")

# Delhi to New Delhi
df.dropna(subset = ['CityLocation'], inplace = True)
df.loc[:,'CityLocation'] = df['CityLocation'].str.replace(r'\bDelhi\b', 'New Delhi')
df.loc[:,'CityLocation'] = df['CityLocation'].str.title()

# Small dataset
df_cities = df[['CityLocation', 'AmountInUSD']]

# group NCR cities
replace_dict = {
    'New Delhi':'NCR',
    'Gurgaon':'NCR',
    'Noida':'NCR'
}
df_cities.loc[:,'CityLocation'] = df_cities['CityLocation'].replace(replace_dict)

cities = ['Bangalore','NCR','Mumbai','Pune','Hyderabad']
df_cities = df_cities[df_cities['CityLocation'].isin(cities)]

df_cities.head()
```

```python
population_size = df_cities.size
sample_size = 100

cities = ['Bangalore','NCR','Mumbai','Pune','Hyderabad']

df_strat = pd.DataFrame()
for i in cities:
    df_temp = df_cities[df_cities['CityLocation'] == i]
    values_per_strata = df_temp.size
    n_points = (values_per_strata / population_size ) * sample_size
    df_temp = df_temp.sample(int(n_points))
    df_strat = pd.concat([df_strat, df_temp], ignore_index=True)
    
avgSample = df_strat['AmountInUSD'].mean()
avgPopulation = df_cities['AmountInUSD'].mean()
samplingError = avgPopulation - avgSample
print('Sample mean ', avgSample)
print('Population mean ', avgPopulation)
print('Sampling Error ', samplingError)
```
